## 4.2.3 データの整形

```python
import pandas as pd
import numpy as np
```

### 使用するデータの読み込み

```python
df = pd.read_excel("data/201704health.xlsx")
df
```

### 条件で抽出

```python
# 10000歩以上の日
df["歩数"] >= 10000 # df.loc[:, "歩数"] >= 10000
```

```python
df_selected = df[df["歩数"] >= 10000]
df_selected
```

```python
df_selected.shape
```

```python
df.query('歩数 >= 10000 and 摂取カロリー <= 1800')
```

### データ型変換

```python
df.dtypes
```

```python
df.loc[:, 'date'] = df.loc[:, '日付'].apply(pd.to_datetime)
```

```python
df.loc[:, 'date']
```

```python
df
```

```python
df.loc[:, "摂取カロリー"] = df.loc[:, "摂取カロリー"].astype(np.float32)
```

```python
df = df.set_index("date")
```

```python
df.head()
```

### 並べ替え

```python
df.sort_values(by="歩数")
```

```python
df.sort_values(by="歩数", ascending=False).head()
```

### 不要なカラムを削除

```python
df = df.drop("日付", axis=1)
```

```python
df.tail()
```

### 組合せデータの挿入

```python
df.loc[:, "歩数/カロリー"] = df.loc[:, "歩数"] / df.loc[:, "摂取カロリー"]
df
```

```python
def exercise_judge(ex):
    if ex <= 3.0:
        return "Low"
    elif 3.0 <= 6.0:
        return "Mid"
    else:
        return "High"
```

```python
df.loc[:, "運動指数"] = df.loc[:, "歩数/カロリー"].apply(exercise_judge)
df
```

```python
df.to_pickle("data/df_201704health.pickle")
```

```python
df_moved = pd.get_dummies(df.loc[:, "運動指数"], prefix="運動")
df_moved
```

```python
df_moved.to_pickle("data/df_201704moved.pickle")
```

## 4.2.4 時系列データ

### １ヶ月分のデータを作る

```python
dates = pd.date_range(start="2017-04-01", end="2017-04-30")
dates
```

```python
np.random.seed(123)
df = pd.DataFrame(np.random.randint(1, 31, 30),
                 index=dates, columns=["乱数"])
df
```

### １年分365日のデータを作る

```python
dates = pd.date_range(start="2017-01-01", periods=365)
dates
```

```python
np.random.seed(123)
df = pd.DataFrame(np.random.randint(1, 31, 365), index=dates, columns=["乱数"])
df
```

### 月平均のデータにする

```python
df.groupby(pd.Grouper(freq='M')).mean()
```

```python
df.loc[:, "乱数"].resample("M").mean()
```

### 複雑な条件のインデックス

```python
# 一年分の土曜日の日付データ
pd.date_range(start="2017-01-01", end="2017-12-31", freq="W-SAT")
```

```python
df_year = pd.DataFrame(df.groupby(pd.Grouper(freq="W-SAT")).sum(), columns=["乱数"])
df_year
```

## 4.2.5 欠損値処理

```python
import pandas as pd
df_201705 = pd.read_csv("data/201705health.csv", encoding="utf-8",
                       index_col='日付', parse_dates=True)
df_201705
```

```python
df_201705_drop = df_201705.dropna() # dropna删除NaN欠损行
df_201705_drop
```

```python
df_201705_fillna = df_201705.fillna(0) # fillna(0)将NaN欠损值换为0
df_201705_fillna
```

```python
df_201705_fill = df_201705.fillna(method='ffill') # 使用前一个值填补欠损值
df_201705_fill
```

```python
# 使用平均值mean()，中值median()，最频值mode().iloc[0, :]
df_201705_fillmean = df_201705.fillna(df_201705.mean())
df_201705_fillmean
```
