# 4.3 matplotlib

#### 日本語でラベルを表示
> matplotlibrcという名前で下記パスに作成する  
> MacOS: ~/.matplot/  
> `font.sans-serif: IPAexGothic, SimHei`
> # 或者使用下面语句

plt.rcParams['axes.unicode_minus'] = False`


```python
# 配置文件地址
import matplotlib
matplotlib.matplotlib_fname()

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['IPAexGothic']
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号
```

```python
import matplotlib.style

# ggplotスタイルの指定
matplotlib.style.use('ggplot')

fig, ax = plt.subplots()

ax.set_title('日本語のタイトルを設定')
ax.text(2, 6, '文字列を描画', size=20)

x = [1, 2, 3]
y = [10, 2, 3]
labels = ['スパム', 'ハム', '玉子']
ax.bar(x, y, tick_label=labels, label='食べ物')

plt.show()
```

## 4.3.3 グラフの種類と出力方法

### 折れ線グラフ

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 1, 2]
ax.plot(x, y1) # 折れ線グラフを描画
ax.plot(x, y2)

plt.show()
```

```python
import numpy as np

x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fit, ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()

plt.show()
```

### 棒グラフ

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]
ax.bar(x, y) # 棒グラフを描画

plt.show()
```

```python
# tick_label 目盛りにラベルを指定
fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]
labels = ['スパム', 'ハム', '玉子']
ax.bar(x, y, tick_label=labels) # ラベルを指定

plt.show()
```

```python
# barh
fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]
labels = ['スパム', 'ハム', '玉子']
ax.barh(x, y, tick_label=labels)
plt.show()
```

```python
fig, ax = plt.subplots()
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

width = 0.4 # 棒グラフの幅を0.4にする
ax.bar(x, y1, width=width, tick_label=labels, label='y1') # 幅を指定して描画

#　幅分ずらして棒グラフを描画
x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label='y2')

ax.legend()
ax.set_title("複数の棒グラフを並べて描画")

plt.show()
```

```python
fig, ax = plt.subplots()
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

# y1とy2を足した値を格納
y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]

# y1とy2を足した高さの棒グラフを描画
ax.bar(x, y_total, tick_label=labels, label='y1')
ax.bar(x, y2, label='y2')

ax.legend()
ax.set_title("積み上げ棒グラフを描画")

plt.show()
```

### 散布図

```python
# scatter
fig, ax = plt.subplots()

# ランダムに５０個要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x, y) # 散布図を描画
plt.show()
```

```python
# marker 
fig, ax = plt.subplots()

# ランダムに５０個要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x[0:10], y[0:10], marker='v', label='triangle down') # 下向き三角
ax.scatter(x[10:20], y[10:20], marker='^', label='triangle up') # 上向き三角
ax.scatter(x[20:30], y[20:30], marker='s', label='square') # 正方形
ax.scatter(x[30:40], y[30:40], marker='*', label='star') # 星型
ax.scatter(x[40:50], y[40:50], marker='x', label='x') # X

ax.legend()

plt.show()
```

### ヒストグラフ （直方图）

```python
# hist
# 正規分布に従うランダムな値
np.random.seed(123)
mu = 100 # 平均値
sigma = 15 # 標準偏差
x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

# ヒストグラフを描画
n, bins, patches = ax.hist(x)
ax.set_title("ビストグラフを描画")
plt.show()
```

```python
for i, num in enumerate(n):
    print('{:.2f} - {:.2f}: {}'.format(bins[i], bins[i + 1], num))
```

```python
fig, ax = plt.subplots()
ax.hist(x, bins=25) # ピンの数を指定して描画
plt.show()
```

```python
fig, ax = plt.subplots()
# 横向きのヒストグラフを描画
ax.hist(x, orientation='horizontal') 
plt.show()
```

```python
# 複数の値
np.random.seed(123)
mu = 100 # 平均値
x0 = np.random.normal(mu, 20, 1000)

# 異なる標準偏差でデータを生成
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']
# 3つのデータのヒストグラフを描画
ax.hist((x0, x1, x2), label=labels)

ax.legend()
plt.show()
```

```python
# stacked = True 積み上げたヒストグラフ
fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']
# 積み上げたヒストグラフを描画
ax.hist((x0, x1, x2), label=labels, stacked=True)
ax.set_title("積み上げたヒストグラフ")
ax.legend()
plt.show()
```
