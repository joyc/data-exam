# 4.3 matplotlib

## 4.3.1 Matplotlibの概要

```python
import matplotlib.pyplot as plt
```

### MATLABスタイル

```python
import matplotlib.style

# ggplotスタイルの指定
matplotlib.style.use('ggplot')
```

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

plt.plot(x, y) # 折れ線グラフを描画
plt.title('MATLAB-style') # グラフにタイトルを設定

plt.show() # グラフを表示
```

### オブジェクト指向スタイル

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

ax.plot(x, y) # 折れ線グラフを描画
ax.set_title('OOP-style') # イトルを設定

plt.show() # グラフを表示
```

## 4.3.2 描画オブジェクト

### 描画オブジェクトとサブプロット

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2) # 2つのサブプロットを配置
plt.show()
```

```python
fig, axes = plt.subplots(2, 2) # 2行２列の4個サブプロットを配置
plt.show()
```

```python
fig, axes = plt.subplots(ncols=2) # １行２列のサブプロットを配置
plt.show()
```

### グラフのスタイル

```python
import matplotlib.style

# スタイルの一覧を表示
print(matplotlib.style.available)
```

```python
# グラフのスタイルにclassicを指定
matplotlib.style.use('classic')

fig, ax = plt.subplots()
ax.plot([1, 2])

plt.show()
```

### タイトル

```python
fig, axes = plt.subplots(ncols=2)

# サブプロットにタイトルを設定
axes[0].set_title('subplot title 0')
axes[1].set_title('subplot title 1')

# 描画オブジェクトにタイトルを設定
fig.suptitle('figure title')

plt.show()
```

### 軸ラベル

```python
fig, ax = plt.subplots()

ax.set_xlabel('x label') # x軸にラベルを設定
ax.set_ylabel('y label') # y軸にラベルを設定

plt.show()
```

### 凡例

```python
fig, ax = plt.subplots()

# 凡例用のラベルを設定
ax.plot([1,2,3], [2, 4, 9], label='legend label')
ax.legend(loc='best') # 凡例を表示

plt.show()
```

```python
fig, ax = plt.subplots()

# 凡例用のラベルを設定
ax.plot([1,2,3], [2, 4, 9], label='legend label')
ax.legend(loc='lower right') # 凡例を表示

plt.show()
```

> 'upper left', 'center', 'center left', 'lower center', 'best'

### ファイルの出力

```python
fig, ax = plt.subplots()
ax.set_title('subplot title')
fig.savefig('sample-figure.png') # png形式で保存
fig.savefig('sample-figure.svg') # svg形式で保存
```
